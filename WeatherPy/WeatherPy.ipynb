# WeatherPy


### Observations:

1- Based on the graph that shows max temperature of cities vs. latitude, cities located in latitude zero (close to the equator) have the highest max temperature. The max temperature decreases directly with the distance from equator.

2- Humidity of cities is affected by many geographical factors such as acssess to the water resources, mountains, cloudiness, latitude, and so on; therefore, no direct relation between the latitude and the humidity of cities is observable. 

3- A similar conclusion can be drawn about the relation betweenthe latitude and the humidity, the cloudiness and the wind speed of cities. However, it seems that the wind speed in the cities close to the eauator is a little bit more than cities closer than to the poles.

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime
import scipy.stats as st

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "result.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"


city_names = []
clouds = []
countries = []
dates = []
humidities = []
latitudes = []
longitudes = []
max_temps = []
wind_speeds = []

for i in range (len(cities)):
    try:
        query_url = f"{base_url}appid={weather_api_key}&units={units}&q={cities[i]}"
        response = requests.get(query_url).json()
        print(f"Processing Record {i+1} | {cities[i]}")
        country = response["sys"]["country"]
        latitude = response["coord"]["lat"]
        longitude = response["coord"]["lon"]
        date = response["dt"]
        max_temp = 1.8*(response["main"]["temp_max"]) + 32
        humidity = response["main"]["humidity"]
        cloudiness = response["clouds"]["all"]
        wind_speed = 2.236936*(response["wind"]["speed"])
        city_names.append(cities[i])
        latitudes.append(latitude)
        longitudes.append(longitude)
        countries.append(country)
        dates.append(date)
        max_temps.append(max_temp)
        humidities.append(humidity)
        clouds.append(cloudiness)
        wind_speeds.append(wind_speed)
        
    except KeyError:
        print("not found the info of the city!")

weather_data = {"City": city_names,
                "Cloudiness": clouds,
                "Country": countries,
                "Date": dates,
                "Humidity": humidities,
                "Lat": latitudes,
                "Lng": longitudes,
                "Max Temp": max_temps,
                "Wind Speeds": wind_speeds}
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_reports = pd.DataFrame(weather_data)
weather_reports[["Max Temp", "Wind Speeds"]] = weather_reports[["Max Temp", "Wind Speeds"]].apply(pd.to_numeric)
weather_reports["Max Temp"] = weather_reports["Max Temp"].map("{:.2f}".format)
weather_reports["Wind Speeds"] = weather_reports["Wind Speeds"].map("{:.2f}".format)
    
weather_reports.to_csv(output_data_file)
weather_reports.head()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames

# Northern DF
northern_df = weather_reports.loc[weather_reports["Lat"] >= 0,:]

# Southern DF
southern_df = weather_reports.loc[weather_reports["Lat"] < 0,:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Run regression

y = northern_df["Max Temp"].astype(float)
x = northern_df["Lat"]

(slope, intercept, _, _, _) = st.linregress(x, y)

# Get regression values
equation = x * slope + intercept

# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

# To add regress line to your plot:
plt.scatter(x, y, facecolor="b", edgecolor="black")
plt.plot(x,equation ,"r")

# To add the equation to your plot:
plt.annotate(line_eq,(20,45),fontsize=11,color="red")

# Make Labels
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.title(f" Max Temperature vs. Latitude of Northern Hemispher Cities ({plot_date})")
plt.grid()

plt.savefig('Images/figure_5.jpg', dpi=100)
plt.show()
```
